```python
import pandas as pd
```

```python
us_babies = pd.read_csv("us_baby_names.csv")
```

```python
us_babies
```

```python
us_babies["Year"]
```

```python
us_babies["Year"] == 2014
```

```python
#loc is used to call by labels
us_babies_2014 = us_babies.loc[us_babies["Year"] == 2014, :]
```

```python
us_babies_2014
```

```python
sorted_names = us_babies_2014.sort_values("Count" , ascending = False)
```

```python
#iloc is used to call by positions
sorted_names.iloc[0:5]
```

```python
state_babies = pd.read_csv("state_baby_names.csv")
```
